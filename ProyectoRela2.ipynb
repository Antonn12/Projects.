```julia
using Pkg

Pkg.add("LinearAlgebra")
Pkg.add("Images")
```

```julia
function Rungen_Kutta_O4(f, Δ, x0, y0)
    x=x0+Δ
    k1=Δ*f(x0,y0)
    
    k2=Δ*f(x0+(Δ/2),y0+(k1/2))
    
    k3=Δ*f(x0+(Δ/2),y0+(k2/2))
    
    k4=Δ*f(x0+Δ,y0+k3)
    m=(k1 + 2*k2 + 2*k3 + k4)/6
    y = y0 + m
    return [x, y, k1, k2, k3, k4, m]
end
    
```

```julia
function aproximacion(f, Δ, x0, y0, y_aprox)
    ite=0 
    while x0<y_aprox
        resultado=Rungen_Kutta_O4(f, Δ, x0, y0)
        ite+=1
    end
        print("K1   = ", resultado[3])
        print("K2   = ", resultado[4])
        print("K3   = ", resultado[5])
        print("K4   = ", resultado[6])
        print("X = ", resultado[1])
        print("Y = ", resultado[2])
    return [resultado[1], resultado[2]]
end
```
